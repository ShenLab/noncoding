```python
import genomepy

#get hg19 reference genome ZERO INDEXED
genomepy.install_genome("hg19", "UCSC", genome_dir="/data/genomes")
g = genomepy.Genome("hg19", genome_dir="/data/genomes")
```

```python
#mapping base pairs for switching strand
bp = {}
bp['A'] = 'T'
bp['G'] = 'C'
bp['T'] = 'A'
bp['C'] = 'G'
bp['N'] = 'N'


def init_bank():
  bank = {}
  seq_lens = [i for i in range(752,802)]

  for seql in seq_lens:
      d = {}
      for j in np.linspace(0,.8,5): #buckets of 0, .2, .4, .6, .8
          d[round(j,1)] = []

      bank[seql] = d
  return bank


def get_gc(seq):
    count = 0
    for bp in seq:
        if bp == 'C' or bp == 'G':
          count += 1
      
    gc_bin = floor(5*count/len(seq))/5
    if gc_bin == 1:
        gc_bin = .8
    
    return gc_bin
  
  
def switch_strand(seq):
    out = ''
    for s in seq[::-1]: #reverse step
        out += bp[s] #match step
    return out
  
  
def get_rand_seq(seql, ref_table, ref_len):    
  while(1):
    #select random transcript
    r = random.randrange(ref_len)
    row = ref_table.iloc[r]

    #select midpoint
    mid = random.randint(row.start-1, row.end-1)

    pad = seql//2
    if seql%2 == 0:
      if mid == row.start:
        mid += 1
      start = mid - pad
      end = mid + pad
    else:
      start = mid - pad
      end = mid + pad + 1

    s = str(g[row.chr][start:end]).upper()
    neg = random.choice([0,1])
    if neg:
      s = switch_strand(s)

    return (s, row.chr, start+1, end+1, neg, row.lnTPM)
```

```python
###randomly generate random seqs###
pos = pd.read_csv('pos_rbp_seqs_1kbp.csv')
cell_line = 'K562'
RBP = cell_line + '.RBFOX2'
df_RBP = pos[pos['RBP'] == RBP]

###ref table files for cell line###
all_ranges_file = cell_line + '_neg_ranges.csv'
all_ranges_table = pd.read_csv(all_ranges_file)

bad_ranges_file = cell_line + '_RBP_bad_ranges.csv'
bad_ranges_table = pd.read_csv(bad_ranges_file)

bad_ranges = bad_ranges_table[bad_ranges_table['RBP'] == RBP].print_ranges.iloc[0].split('-')
bad_ranges = [int(x)-1 for x in bad_ranges]
ref_table = all_ranges_table.loc[~all_ranges_table.index.isin(bad_ranges)]
ref_len = len(ref_table)
bank = init_bank()

neg_seqs = [0] * len(df_RBP)
count = 0
for seq in df_RBP['sequence']:
  s = get_rand_seq(151, ref_table, ref_len)
  neg_seqs[count] = s
  count += 1

df_neg = pd.DataFrame(neg_seqs, columns=['sequence', 'chr', 'start', 'end', 'strand', 'lnTPM'])
name = cell_line + '_RBFOX2_total_random_neg.csv'
df_neg.to_csv(name)
```
